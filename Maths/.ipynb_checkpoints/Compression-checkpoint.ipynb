# Compression

```julia
using Colors, Plots, LinearAlgebra

function buildW(N)
    W = zeros(N,N)
    W[:,1] = ones(N)/sqrt(N)
    j = (0:N-1)

    for k=2:N
        W[:,k] = cos.((j.+ .5)*(k-1)*π/N)*sqrt(2/N)
    end
    return W
end

function grayW(W)
    N = size(W,1)
    gW = copy(W)
    gW = (W.+sqrt(2/N))*sqrt(N/2)/2
    return gW
end

N = 8
W = buildW(N)
gW = grayW(W)
Gray.(gW)
# typeof(gW[:,1])
# Gray.(gW[:,8]')'
```

```julia
[dot(W[:,k],W[:,l]) for k in 1:N, l in 1:N]
```

```julia
function W2d(N)
    V = Array{Array{Float64, 2}, 2}(undef, N,N)
    W = buildW(N)
    for k in 1:N
        for l in 1:N
            V[k,l] = W[:,k]*W[:,l]'
        end
    end
    return V
end
            

function grayW2d(W2)
    N = size(W2,1)
    gW2 = copy(W2)
    for k in 1:N
        for l in 1:N
            gW2[k,l] = (W2[k,l].+2/N)*N/4
        end
    end
    return gW2
end

W2 = W2d(N)
gW2 = grayW2d(W2);

using Interact:@manipulate
@manipulate for k in 1:N , l in 1:N
       Gray.(gW2[k,l])
end 
```

```julia
function DCT(patch)
    N = size(patch,1)
    W2 = W2d(N)
    coeff = [dot(patch,W2[k,l]) for k in 1:N, l in 1:N]
    return coeff
end

function invDCT(coeff)
    N = size(coeff,1)
    W2 = W2d(N)
    return sum(coeff[k,l]*W2[k,l] for k in 1:N, l in 1:N)
end 
```

```julia
using TestImages, Images
img = testimage("cameraman")
function imblocksView(img,s::Integer)
    n,m = size(img).÷ s
    [view(img, (1:s) .+ (i*s), (1:s) .+ (j*s)) for i=0:n-1, j=0:m-1]
end

imblocksView(img, 32)
```

```julia
function imblocks(img,s::Integer)
    n,m = size(img).÷ s
    patches = [Float32.(img[(1:s) .+ (i*s), (1:s) .+ (j*s)]) for i=0:n-1, j=0:m-1]
    return patches
end
```

```julia
function DCTfull(img,N)
    patches = imblocks(img,N)
    n = size(patches,1)
    Coeff= [DCT(patches[i,j]) for i=1:n, j=1:n]
    return Coeff
end

function reconstruct_img(C)
    N = size(C[1,1],1)
    n = size(C,1)
    return [invDCT(C[i,j]) for i=1:n, j=1:n]
end
```

```julia
function imDeblock(P)
    n = size(P,1)
    s = size(P[1,1],1)
    A = zeros(n*s,n*s)
    for i in 0:n-1
        for j in 0:n-1
            A[(1:s).+(i*s),(1:s) .+ (j*s)] = P[i+1,j+1]
        end
    end
    return A
end
```

```julia
using Interact:@manipulate,slider

N = 16

C = DCTfull(img,N)

Gray.(abs.(imDeblock(C)))
```

```julia
n,m = size(C)

@manipulate for s=slider(0:.002:1, value=0)

    Cnew = copy(C)
    nonzeroCoeffs = 0
    for i in 1:n
        for j in 1:m
            Cnew[i,j] = C[i,j].*(abs.(C[i,j]).>s)
            nonzeroCoeffs = nonzeroCoeffs + length(Cnew[i,j][abs.(Cnew[i,j]).>0])
        end
    end
    
@show nonzeroCoeffs    
    
    Gray.(imDeblock(reconstruct_img(Cnew)))
end
```
