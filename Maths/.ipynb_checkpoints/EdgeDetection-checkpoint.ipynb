# Prewitt / Sobel edge detection

```julia
using ImageFiltering, TestImages, ImageShow, Colors

img1 = testimage("cameraman")
img2 = testimage("resolution_test_1920")
img3 = Gray.(testimage("peppers_gray"))
img4 = Gray.(testimage("sudoku"))


display(img1)
display(img2)
display(img3)
display(img4)
#Gray.(img4)
```

```julia
function mySmoothing(img,n) 
    G = Kernel.gaussian(n)
    return imfilter(img,G)
end

display(mySmoothing(img1,5))
display(mySmoothing(img2,4))
display(mySmoothing(img3,3))
display(mySmoothing(img4,1))
```

```julia
prewitt_x, prewitt_y = Kernel.prewitt()
sobel_x , sobel_y = Kernel.sobel()
display(6*prewitt_x)
display(8*sobel_x)

display(6*prewitt_y)
display(8*sobel_y)
```

```julia
function myPrewitt(img)
    
    prewitt_x, prewitt_y = Kernel.prewitt()
    
    imgx = imfilter(img,prewitt_x)
    imgy = imfilter(img,prewitt_y)
    
    return sqrt.(imgx.^2 .+ imgy.^2)
end

function mySobel(img)
    
    sobel_x , sobel_y = Kernel.sobel()

    imgx = imfilter(img,sobel_x)
    imgy = imfilter(img,sobel_y)

    return sqrt.(imgx.^2 .+ imgy.^2)
end


E1P = myPrewitt(img1)
E1S = mySobel(img1)
Gray.(E1S)
```

```julia
E2P = myPrewitt(img2)
E2S = mySobel(img2)
Gray.(E2S)
```

```julia
E3P = myPrewitt(img3)
E3S = mySobel(img3)
Gray.(E3S)
```

```julia
E4P = myPrewitt(img4)
E4S = mySobel(img4)
Gray.(E4S)
```

```julia
using Interact:@manipulate,slider

EE1 = Float64.(E1S)

@manipulate for s=slider(0.01:.01:.5, value=0.01)
    T1 = EE1.>s
    Gray.(T1)
end
```

```julia
EE2 = Float64.(E2S)

@manipulate for s=slider(0.01:.01:.5, value=0.01)
    Gray.(EE2.>s)
end
```

```julia
EE3 = Float64.(E3S)

@manipulate for s=slider(0.01:.01:.5, value=0.01)
    Gray.(EE3.>s)
end
```

```julia
EE4 = Float64.(E4S)

@manipulate for s=slider(0.01:.01:.5, value=0.01)
    Gray.(EE4.>s)
end
```
