```julia
using Colors, ImageShow, Images, TestImages, Statistics
img = testimage("cameraman")
```

```julia
σ = 0.1 
S = round(Int,length(img)*σ)

noisy_img = copy(img)

noisy_img[rand(1:length(img), S)] .= 0
noisy_img[rand(1:length(img), S)] .= 1



using Interact:slider,@manipulate
@manipulate for s = slider(1:2:21, value=1)
    denoised_img = mapwindow(median!, noisy_img, (s, s))
    mosaicview(noisy_img, denoised_img; nrow=1)
end 
```
