### Subtask 1

In this Task the Lloyd algorithm has to be implemented to determine the different clusters of data. In this Subtask,  we load define and visualize said data according to the provieded code snippets. The four different data sets to work with are plotted below.

```python
import numpy as np
from sklearn.datasets import make_blobs

n_samples = 1500
random_state = 170
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]

X, y = make_blobs(n_samples=n_samples, random_state=random_state)

X_aniso = np.dot(X, transformation) # Anisotropic blobs

X_varied, y_varied = make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=
    random_state
) # Unequal variance

X_filtered = np.vstack(
    (X[y == 0][:500], X[y == 1][:100], X[y == 2][:10])
) # Unevenly sized blobs

y_filtered = [0] * 500 + [1] * 100 + [2] * 10
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

axs[0, 0].scatter(X[:, 0], X[:, 1], c=y)
axs[0, 0].set_title("Mixture of Gaussian Blobs")

axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=y)
axs[0, 1].set_title("Anisotropically Distributed Blobs")

axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=y_varied)
axs[1, 0].set_title("Unequal Variance")

axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_filtered)
axs[1, 1].set_title("Unevenly Sized Blobs")

plt.suptitle("Ground truth clusters").set_y(0.95)
plt.show()
```

### Subtask 2

As mentioned above, in this Subtask we had to implement the Lloyd algorithm ourselves. For this, I first defined the "myCluster" class, which stores the center and the data points in its periphery and helps with calculating the mean point as well as the reslpective objective. The algorithm itself was split into three methods, where the "lloyd_algorithm_step(...)" and "update_center_step(...)" were defined separately to make it more readable. As convergence criterion I used $0.0001$, but this can be specified differently with the parameter "convergence_value" in the "lloyds_algorithm"- function. 

The algorithm was used for the values $k=\{1,2,3,4,5,6\}$. The plots below show the different outcomes for these k-values and the corresponding change in the objective for each of these datasets.

For most datasets the "best visual" result considering the three clusters given in the assignment is either for $k=2$ (eg. for the unevenly sized blobs) or $k=3$ (e.g. gaussian blobs).  Alhough, the anisotropically distributed blobs cannot be fully separated since we are using centrical distances. For this dataset a kernel metric would probably achieve better results.

This observation also corresponds to the plot of the objectives, which decreases most in the first few k-values. Especially, for the unevenly sized blobs, where the first two values have a huge difference whereas the rest changes much less and rather evenly. Also interesting to see is that the cluster with the least amount of data points has its own center mostly if $k\geq 4$. But this also greatly depends on the random points that are chosen as centers. The rest of the additional centers are collectively part of the cluster with the most points.

For the data with different variances, most of the centers for $k>3$ will be a part of the one cluster with the biggest variance, since thise points are spread out and additional centers decrease the minimum distances more than thy would for the other clusters.

```python
#set convergence and plot sizes(only for those with subplots) for rest of tasks
convergence_crit=1./10000
figure_size_all=(14, 10)
```

```python
import math
def euclid_distance(point_a, point_b):
    #return math.sqrt((point_a[0] - point_b[0])**2 + (point_a[1] - point_b[1])**2)**2
    return np.linalg.norm(point_a - point_b)**2

#print(euclid_distance(np.array([0,0]),np.array([0,3])))
```

```python
class myCluster:
    
    def __init__(self,center):
        self.center_point= center
        self.points= []
        
    def isEmpty(self):
        if len(self.points)==0:
            return True
        else:
            return False
    
    def calc_objective(self):
        sum= 0
        for x in self.points:
            sum+= euclid_distance(self.center_point,x)
        return sum
    
    def calc_mean(self):
        sum= np.array([0,0],dtype=np.float64)
        for x in self.points:
            sum+=x
        return sum/len(self.points)
    
    def add_point(self,point):
        if self.isEmpty():
            self.points=[point]
        else:
            self.points=np.append(self.points,[point],axis=0)
        
        
    def clear_and_update_center(self,new_center):
        self.center_point= new_center
        self.points=[]
        
```

```python
import random
def lloyds_algorithm(dataset, k_clusters,convergence_value, random_init):
    centers=[]
    clusters=[]
    objectives=np.array([],dtype=np.float64)
    if(random_init):
        #random centers
        for k in range(0,k_clusters):
            random_center_index= random.randint(0, len(dataset))
            centers.append(dataset[random_center_index,:])
            clusters.append(myCluster(dataset[random_center_index,:]))
    else:
        #for subtask 3
        centers, clusters = kpp_init(dataset.copy(),k_clusters)
    for c in clusters:
        objectives=np.append(objectives,c.calc_objective())
    difference= objectives
    #range to ensure an ending
    for step in range(0,1000):
        filled_clusters= lloyd_algorithm_step(dataset.copy(),clusters.copy())
        new_objectives = np.array([],dtype=np.float64)
        for c in clusters:
            new_objectives=np.append(new_objectives,c.calc_objective())
        difference_in_objectives= objectives-new_objectives
        if all(np.linalg.norm(i-j) <=  convergence_value for i,j in zip(difference,difference_in_objectives)):
            print("ending algorithm for k={} at step {} due to convergence".format(k_clusters,step))
            return filled_clusters
        difference=difference_in_objectives
        clusters=update_center_step(dataset.copy(),filled_clusters.copy())
    return clusters

def lloyd_algorithm_step(dataset_copy, clusters_to_fill):
    #finding nearest center& filling the corresponding cluster
    filling_clusters=clusters_to_fill.copy()
    for d in range(0,len(dataset_copy)):
        all_distances=[]
        for cl in filling_clusters:
            current_center=cl.center_point
            all_distances.append(euclid_distance(current_center,dataset_copy[d,:]))
        index_min_dist=all_distances.index(min(all_distances))
        filling_clusters[index_min_dist].add_point(dataset_copy[d,:])
    return filling_clusters
    

def update_center_step(dataset_copy, clusters_to_clear):
    #the centers are adapted to the mean or a random new one and the points in the clusters are cleared
    step_objectives= []
    for cl in clusters_to_clear:
        step_objectives.append(cl.calc_objective())
        if cl.isEmpty():
            random_center_index= random.randint(0, len(dataset_copy))
            cl.clear_and_update_center(dataset_copy[random_center_index,:])
        else :
            mean_center= cl.calc_mean()
            cl.clear_and_update_center(mean_center)
    return clusters_to_clear
```

```python
def visualize_lloyd_clusters(all_clusters,axis,axs_field):
    #creating subplots for each k
    colors=['cyan','r','y','g','c','m']
    fields=[[0,0],[0,1],[0,2],[1,0],[1,1],[1,2]]
    for index in range(0,len(all_clusters)):
        list_points = all_clusters[index].points
        center=all_clusters[index].center_point
        axis[fields[axs_field][0],fields[axs_field][1]].scatter(list_points[:,0], list_points[:,1], s=10, c=colors[index], label="cluster "+str(index+1))
        axis[fields[axs_field][0],fields[axs_field][1]].scatter(center[0],center[1],s=50,c='k', marker="X")
        axis[fields[axs_field][0],fields[axs_field][1]].set_title(str(axs_field+1)+" clusters")
    
```

```python
k_values=[1,2,3,4,5,6]
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=figure_size_all)

objective_sums=[]
for k in k_values:
    current_clusters= lloyds_algorithm(X, k, convergence_crit,True)
    objectives=np.array([],dtype=np.float64)
    for c in current_clusters:
        objectives=np.append(objectives,c.calc_objective())
    if k==1:
        objective_sums=[[k,np.sum(objectives)]]
    else:
        objective_sums=np.append(objective_sums,[[k,np.sum(objectives)]],axis=0)
    visualize_lloyd_clusters(current_clusters,axs, k-1)
    
plt.suptitle("Mixture of Gaussian Blobs").set_y(0.95)
plt.show()

print(objective_sums)
fig, ax = plt.subplots(figsize=(7,5))
plt.plot(objective_sums[:,0],objective_sums[:,1])
plt.title("Objective function values for Mixture of Gaussian Blobs")
plt.show()
```

### Subtask 3

In the third subtask the initial cluster centers should be derived via the kmeans++ algorithm. The following plots show the results of the same datasets and k-values as above.

The biggest visible change through the plots of the clusters, is that the smallest cluster in the unevenly sized dataset is more often found with also smaller values, once even with $k=2$. But seeing as there are not as many changes as I initially thought there would be, my algorithm for kmeans++ could probably be enhanced to get better and faster results.

```python
#centers, clusters = kpp_init(dataset.copy(),k_clusters)
def kpp_init(dataset_pp,k_clusters):
    centerspp=[]
    clusterspp=[]
    objectivespp=np.array([],dtype=np.float64)
    #first center random
    random_center_index= random.randint(0, len(dataset_pp))
    first_center_random=dataset_pp[random_center_index,:]
    centerspp.append(first_center_random)
    clusterspp.append(myCluster(first_center_random))
    
    #other depending on distance to already existing centers
    for i_k in range(1,k_clusters):
        all_min_distances=[]
        intermediate_objective=0.
        for index in range(0,len(dataset_pp)):
            possible_distances=[]
            for cl in clusterspp:
                current_center=cl.center_point
                possible_distances.append(euclid_distance(current_center,dataset_pp[index,:]))
            intermediate_objective+=min(possible_distances)
            all_min_distances.append(min(possible_distances))
        rand_probability= (random.randint(0, 100)/100)*intermediate_objective
        sum_distance=0.
        possible_center_index= 0
        
        while sum_distance< rand_probability and possible_center_index<len(dataset_pp)-1:
            sum_distance+=all_min_distances[possible_center_index]
            possible_center_index+=1
        
        centerspp.append(dataset_pp[possible_center_index,:])
        clusterspp.append(myCluster(dataset_pp[possible_center_index,:]))
    return centerspp, clusterspp
    
```
