# Task 2: MLP Classifier

```python
# all imports
import numpy as np
import matplotlib.pyplot as plt
import time

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

from stochastic_optimizers import AdamOptimizer

import h5py
```

## Subtask 1: Classification Data

```python
# loading in the train and test sets
hf = h5py.File('classification.h5', 'r')
x_train = np.array(hf.get('x_train'))
y_train = np.array(hf.get('y_train'))
x_test = np.array(hf.get('x_test'))
y_test = np.array(hf.get('y_test'))
hf.close()

# normalize the data using MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)

# set the seed
np.random.seed(42)

# dataset statistics
num_classes = np.unique(y_train).size
num_features = x_train.shape[1]
num_samples_train = x_train.shape[0]
num_samples_test = x_test.shape[0]

print(f"Number of Classes: {num_classes}")
print(f"Number of Features: {num_features}")
print(f"Number of Training Samples: {num_samples_train}")
print(f"Number of Test Samples: {num_samples_test}")
```

## Subtask 2: Forward Propagation


```python

class MLP:
    """
    Multi-Layer Perceptron (MLP) neural network.

    Parameters:
    - input_size (int): Number of input features.
    - hidden_layers_num (int): Number of hidden layers.
    - hidden_size (int): Number of neurons in each hidden layer.
    - output_size (int): Number of output neurons.
    - activation_function (function): Activation function for hidden layers.

    Methods:
    - forward(x): Forward pass through the network.
    - compute_loss(predictions, targets): Compute the perceptron loss.
    - finite_difference_gradients(x, y, epsilon=1e-6): Compute finite difference gradients for weights and biases.
    - train(x_train, y_train, x_test, y_test, epochs, learning_rate): Train the network using finite difference gradients.
    """

    def __init__(self, input_size, hidden_layers_num, hidden_size, output_size, activation_function):
        """
        Initialize weights and biases for the network.

        Parameters:
        - input_size (int): Number of input features.
        - hidden_layers_num (int): Number of hidden layers.
        - hidden_size (int): Number of neurons in each hidden layer.
        - output_size (int): Number of output neurons.
        - activation_function (function): Activation function for hidden layers.
        """
        self.hidden_layers_num = hidden_layers_num
        self.activation_function = activation_function
        self.weights = [np.random.randn(input_size, hidden_size)]
        self.biases = [np.zeros(hidden_size)]

        for _ in range(hidden_layers_num - 1):
            self.weights.append(np.random.randn(hidden_size, hidden_size))
            self.biases.append(np.zeros(hidden_size))

        self.weights.append(np.random.randn(hidden_size, output_size))
        self.biases.append(np.zeros(output_size))
    
    def forward(self, x):
        """
        Forward pass through the network.

        Parameters:
        - x (numpy.ndarray): Input data.

        Returns:
        - numpy.ndarray: Output of the network.
        """
        self.layers = [self.activation_function(np.dot(x, self.weights[0]) + self.biases[0])]
        
        for i in range(1, self.hidden_layers_num + 1):
            layer = self.activation_function(np.dot(self.layers[i-1], self.weights[i]) + self.biases[i])
            self.layers.append(layer)

        return self.layers[-1]
    
    def compute_loss(self, predictions, targets):
        """
        Compute the perceptron loss.

        Parameters:
        - predictions (numpy.ndarray): Model predictions.
        - targets (numpy.ndarray): True labels.

        Returns:
        - float: Mean squared error loss.
        """
        return np.mean((predictions - targets) ** 2)

    def finite_difference_gradients(self, x, y, epsilon=1e-6):
        """
        Compute finite difference gradients for weights and biases.

        Parameters:
        - x (numpy.ndarray): Input data.
        - y (numpy.ndarray): True labels.
        - epsilon (float): Small value for finite difference approximation.

        Returns:
        - list of numpy.ndarray: Gradients for weights.
        - list of numpy.ndarray: Gradients for biases.
        """
        gradients_weights = [np.zeros_like(w) for w in self.weights]
        gradients_biases = [np.zeros_like(b) for b in self.biases]

        for i in range(self.hidden_layers_num + 1):
            for j in range(len(self.weights[i])):
                for k in range(len(self.weights[i][j])):
                    # Compute the finite difference approximation for weights
                    self.weights[i][j][k] += epsilon
                    predictions_plus = self.forward(x)
                    loss_plus = self.compute_loss(predictions_plus, y)

                    self.weights[i][j][k] -= 2 * epsilon
                    predictions_minus = self.forward(x)
                    loss_minus = self.compute_loss(predictions_minus, y)

                    gradients_weights[i][j][k] = (loss_plus - loss_minus) / (2 * epsilon)

                    # Reset the weight for the next iteration
                    self.weights[i][j][k] += epsilon
                    
            for j in range(self.weights[i].shape[1]):
                # Compute the finite difference approximation for biases
                self.biases[i][j] += epsilon
                predictions_plus = self.forward(x)
                loss_plus = self.compute_loss(predictions_plus, y)

                self.biases[i][j] -= 2 * epsilon
                predictions_minus = self.forward(x)
                loss_minus = self.compute_loss(predictions_minus, y)

                gradients_biases[i][j] = (loss_plus - loss_minus) / (2 * epsilon)

                # Reset the bias for the next iteration
                self.biases[i][j] += epsilon

        return gradients_weights, gradients_biases

    def train(self, x_train, y_train, x_test, y_test, epochs, learning_rate):
        """
        Train the network using finite difference gradients.

        Parameters:
        - x_train (numpy.ndarray): Training data.
        - y_train (numpy.ndarray): Training labels.
        - x_test (numpy.ndarray): Testing data.
        - y_test (numpy.ndarray): Testing labels.
        - epochs (int): Number of training epochs.
        - learning_rate (float): Learning rate for optimization.

        Returns:
        - tuple of lists: Training and testing losses and accuracies over epochs.
        """
        optimizer = AdamOptimizer(self.weights + self.biases, learning_rate, 0.9, 0.999, 1e-08)

        train_losses = []
        test_losses = []
        train_accuracies = []
        test_accuracies = []

        for epoch in range(epochs):
            predictions_train = np.clip(np.round(self.forward(x_train)), 0, 1)
            loss_train = self.compute_loss(predictions_train, y_train)

            gradients_weights, gradients_biases = self.finite_difference_gradients(x_train, y_train)
            optimizer.update_params(self.weights + self.biases, gradients_weights + gradients_biases)

            predictions_test = np.clip(np.round(self.forward(x_test)), 0, 1)
            loss_test = self.compute_loss(predictions_test, y_test)

            train_accuracy = accuracy_score(y_train, predictions_train)
            test_accuracy = accuracy_score(y_test, predictions_test)

            train_losses.append(loss_train)
            test_losses.append(loss_test)
            train_accuracies.append(train_accuracy)
            test_accuracies.append(test_accuracy)

            if epoch % 100 == 0:
                print(f"Epoch: {epoch}, Training Loss: {loss_train}, Test Loss: {loss_test}, "
                      f"Training Accuracy: {train_accuracy}, Test Accuracy: {test_accuracy}")

        return train_losses, test_losses, train_accuracies, test_accuracies
```

My implementaition creates a Multi-Layer Perceptron using NumPy and Adam optimization. The network is initialized with parameters input size, number of hidden layers, hidden layer size, output size, and an activation function. Weights and biases are randomly initialized. They are stored as arrays of arrays as parameters of a class instance.

The forward method computes the output of the network for a given input. The compute_loss method calculates the perceptron loss. The finite_difference_gradients method approximates gradients using finite differences for both weights and biases as seen in the task description.

For training, the network uses the Adam optimizer to update weights and biases. During each epoch, it there is a forward pass, loss computing, gradients calculation and parameters updates. The training progress is printed every 100 epochs.


## Subtask 3: Finite Difference Gradient & Optimization
Initialize 4 different models using 2 different architectures and activation functions

```python
# constants
input_size = x_train.shape[1]
output_size = 1

# initialize the relu function as a possible activation function
def relu(x):
    return np.maximum(0, x)

# initialize the sigmoid function as a possible activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# the 4 models
mlp1 = MLP(input_size=input_size, hidden_layers_num=2, hidden_size=20, output_size=output_size, activation_function=relu)
mlp2 = MLP(input_size=input_size, hidden_layers_num=2, hidden_size=20, output_size=output_size, activation_function=sigmoid)
mlp3 = MLP(input_size=input_size, hidden_layers_num=3, hidden_size=8, output_size=output_size, activation_function=relu)
mlp4 = MLP(input_size=input_size, hidden_layers_num=3, hidden_size=8, output_size=output_size, activation_function=sigmoid)
```

Training and reporting the different models

```python
# constans
epochs = 2000
learning_rate = 0.001
```

```python
# iterating through the 4 models
for mlp in [mlp1, mlp2, mlp3, mlp4]:

    # training the first MLP
    start_time = time.time()
    train_losses, test_losses, train_accuracies, test_accuracies = mlp.train(
        x_train,
        y_train,
        x_test,
        y_test,
        epochs,
        learning_rate
    )
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Training Time: {elapsed_time} seconds")

    # plotting mean perceptron loss
    plt.plot(range(epochs), train_losses, label='Training Loss')
    plt.plot(range(epochs), test_losses, label='Test Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Mean Perceptron Loss')
    plt.legend()
    plt.show()

    # plotting accuracy
    plt.plot(range(epochs), train_accuracies, label='Training Accuracy')
    plt.plot(range(epochs), test_accuracies, label='Test Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

The required measures are reported above.
Differences in performance for the different models can be observed.
First of all, the training times tend to differ in every case: 744 seconds, 1359 seconds, 212 seconds and 380 seconds.
Secondly, when using the sigmoid function, the models get stuck in a local minimum and never move.
In the case of the relu function the models converge to a test accuracy of 60%.
