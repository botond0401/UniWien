# Task 1: Non-linear Regression and Classification
## Subtask 1: Toy Data Regression

```python
import h5py
import numpy as np

hf = h5py.File('toy-regression.h5', 'r')
x_train = np.array(hf.get('x_train'))
y_train = np.array(hf.get('y_train'))
x_test = np.array(hf.get('x_test'))
y_test = np.array(hf.get('y_test'))
hf.close()
```

```python
print(f'Number of features: {len(x_train[0])}')
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

surf = ax.scatter([x[0] for x in x_train], [x[1] for x in x_train], y_train)

ax.set_xlabel('first feature')
ax.set_ylabel('second feature')
ax.set_zlabel('output')
fig = plt.figure()
ax = fig.add_subplot()

surf = ax.scatter([x[0] for x in x_train], y_train, cmap='viridis')

ax.set_xlabel('first feature')
ax.set_ylabel('output')
plt.show()

fig = plt.figure()
ax = fig.add_subplot()

surf = ax.scatter([x[1] for x in x_train], y_train, cmap='viridis')

ax.set_xlabel('second feature')
ax.set_ylabel('output')
plt.show()
```

```python
def add_features(array):
    first_feature = array[:, 0]
    second_feature = array[:, 1]
    first_abs = 5 * np.abs(first_feature - 1)
    second_abs = 5 * np.abs(second_feature - 2)
    product = first_abs * second_abs
    length = np.sqrt(first_abs**2 + second_abs**2)
    length2 = np.sqrt(np.abs(first_abs**3) + np.abs(second_abs**3))
    modified_array = np.column_stack((first_feature, second_feature, first_abs, second_abs, product, length, length2))
    
    return modified_array
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

x_train_modified = add_features(x_train)
x_test_modified = add_features(x_test)

model = LinearRegression()
model.fit(x_train_modified, y_train)

y_pred = model.predict(x_test_modified)
mse = mean_squared_error(y_test, y_pred)

print(f'MSE: {mse}')
```

After plotting each features individually I realized that an absolute value function describes really well the target variable. Hence I constructed these absolute value features, but as they were not enough to have the required MSE threshold, I also added some functions of these absolute value features.

After adding some I was able to create a model with an MSE on the test set under 0.01 and my features are also not polynomial.

```python
from sklearn import preprocessing

mse_degrees_test = []
mse_degrees_train = []

degrees = [i for i in range(2, 9)]

for d in degrees:
    poly = preprocessing.PolynomialFeatures(degree=d)
    poly.fit(x_train)
    x_train_poly = poly.transform(x_train)
    x_test_poly = poly.transform(x_test)
    
    model = LinearRegression()
    model.fit(x_train_poly, y_train)
    
    y_pred_train = model.predict(x_train_poly)
    mse_train = mean_squared_error(y_train, y_pred_train)
    mse_degrees_train.append(mse_train)
    
    y_pred = model.predict(x_test_poly)
    mse_test = mean_squared_error(y_test, y_pred)
    mse_degrees_test.append(mse_test)

plt.plot(degrees, mse_degrees_train, 'b', label='Training set')
plt.plot(degrees, mse_degrees_test, 'r', label='Test set')
plt.xlabel("Degree of polynomial")
plt.ylabel("MSE")
plt.legend()
plt.grid()
plt.show()

print(f'The best performance is {min(mse_degrees_test)} with degree {2 + mse_degrees_test.index(min(mse_degrees_test))}.')
```

## Subtask 2: Toy Data Classification

```python
hf2 = h5py.File('toy-classification.h5', 'r')
x_train = np.array(hf2.get('x_train'))
y_train = np.array(hf2.get('y_train'))
x_test = np.array(hf2.get('x_test'))
y_test = np.array(hf2.get('y_test'))
hf2.close()
```

```python
print(f'Number of features: {len(x_train[0])}')
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

surf = ax.scatter([x[0] for x in x_train], [x[1] for x in x_train], y_train)

ax.set_xlabel('first feature')
ax.set_ylabel('second feature')
ax.set_zlabel('output')
fig = plt.figure()
ax = fig.add_subplot()

surf = ax.scatter([x[0] for x in x_train], y_train, cmap='viridis')

ax.set_xlabel('first feature')
ax.set_ylabel('output')
plt.show()

fig = plt.figure()
ax = fig.add_subplot()

surf = ax.scatter([x[1] for x in x_train], y_train, cmap='viridis')

ax.set_xlabel('second feature')
ax.set_ylabel('output')
plt.show()
```

```python
def add_features(array):
    first_feature = array[:, 0]
    second_feature = array[:, 1]
    first_abs = np.abs(first_feature) - 1
    second_abs = np.abs(second_feature) - 1
    product = first_abs * second_abs
    length = np.sqrt(first_abs**2 + second_abs**2)
    length2 = np.sqrt(np.abs(first_abs**3) + np.abs(second_abs**3))
    length3 = np.sqrt(np.abs(first_abs**4) + np.abs(second_abs**4))
    sin1 = np.sin(first_abs)
    sin2 = np.sin(second_abs)
    sin12 = np.sin(first_abs)**2
    sin22 = np.sin(second_abs)**2
    exp1 = np.exp(first_abs)
    exp2 = np.exp(second_abs)
    exp_product = exp1 * exp2
    modified_array = np.column_stack((first_feature, second_feature, first_abs, second_abs, product, length,
                                      length3, sin1, sin2, sin12, sin22, exp1, exp2, exp_product))

    return modified_array
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

x_train_modified = add_features(x_train)
x_test_modified = add_features(x_test)

model = LinearSVC(max_iter=10000)
model.fit(x_train_modified, y_train)
y_pred = model.predict(x_test_modified)

acc = accuracy_score(y_test, y_pred)
print(f"Accuracy of the model: {acc}")
```

My approach was the same as before: after plotting I realized an absolute value function shaped connection, so I created the first features, but as they were not enough, I added different functions of these first features until my model achieved the required 95%.

```python
from sklearn.metrics import confusion_matrix

C = confusion_matrix(y_test, y_pred)
print(C)
```

Looking at the confusion matrix one can observe that model does not make huge mistakes, the differnce between the prediction and the actual value is always 0 or 1. The hardest task for the model is to predict observations with target variable of 4, it only correctly predicts 8 out of 14. The reason of this is probably that compared to the other classes, there are only a few observations in this one.

## Subtask 2: Implement Cross-validation

```python
def run_m_c_cross_validation(n, x, y):

    mse = []

    for _ in range(n):
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)

        model = LinearRegression()
        model.fit(x_train, y_train)

        y_pred = model.predict(x_test)
        mse.append(mean_squared_error(y_test, y_pred))

    return sum(mse)/n
```

```python
mse = run_m_c_cross_validation(100, x, y)
print(f'The estimated MSE with Monte Carlo cross-validation is: {mse}')
```

```python
import math

def create_folds(k, x, y):
    subset_size = math.ceil(len(x) / k)
    data_subsets = []

    for i in range(k-1):
        start = i * subset_size
        end = (i + 1) * subset_size
        x_test_subset = x[start:end]
        y_test_subset = y[start:end]
        x_train_subset = np.concatenate((x[:start], x[end:]))
        y_train_subset = np.concatenate((y[:start], y[end:]))
        data_subsets.append((x_train_subset, x_test_subset, y_train_subset, y_test_subset))

    x_test_subset = x[end:]
    y_test_subset = y[end:]
    x_train_subset = x[:end]
    y_train_subset = y[:end]
    
    data_subsets.append((x_train_subset, x_test_subset, y_train_subset, y_test_subset))
    
    return data_subsets
```

```python
def run_k_f_cross_validation(k, x, y):
    sample_size = len(x)
    shuffled_indices = np.arange(sample_size)
    np.random.shuffle(shuffled_indices)
    x = x[shuffled_indices]
    y = y[shuffled_indices]
    data_subsets = create_folds(k, x, y)
    mse = []
    for x_train, x_test, y_train, y_test in data_subsets:
        model = LinearRegression()
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        mse.append(mean_squared_error(y_test, y_pred))
    return sum(mse)/k
    
```

```python
mse = run_k_f_cross_validation(10, x, y)
print(f'The estimated MSE with Monte Carlo cross-validation is: {mse}')
```

Both MSE are really close to the one from one-time linear regression model, but naturally way better, than the polynomial model's MSE.

```python
mses = []
for _ in range(20):
    mse = run_m_c_cross_validation(40, x, y)
    mses.append(mse)
mean = sum(mses) / len(mses)
diff = [(x - mean) ** 2 for x in mses]
var = sum(diff) / len(diff)
    
print(f'The variance of the MSEs with Monte Carlo cross-validation is: {var}')
```

```python
mses = []
for _ in range(20):
    mse = run_k_f_cross_validation(10, x, y)
    mses.append(mse)
mean = sum(mses) / len(mses)
diff = [(x - mean) ** 2 for x in mses]
var = sum(diff) / len(diff)
    
print(f'The variance of the MSEs with k-fold cross-validations of 10 folds is: {var}')
```

```python
mses = []
for _ in range(20):
    mse = run_k_f_cross_validation(20, x, y)
    mses.append(mse)
mean = sum(mses) / len(mses)
diff = [(x - mean) ** 2 for x in mses]
var = sum(diff) / len(diff)
    
print(f'The variance of the MSEs with k-fold cross-validations of 20 folds is: {var}')
```

The difference between the variances are not significant.
## Subtask 3: Model Selection
### (1)

```python
best_degrees = []
```

```python
mse_degrees_train = []
degrees = [i for i in range(1, 6)]

for d in degrees:
    poly = preprocessing.PolynomialFeatures(degree=d)
    poly.fit(x_train)
    x_train_poly = poly.transform(x_train)
    
    model = LinearRegression()
    model.fit(x_train_poly, y_train)
    
    y_pred_train = model.predict(x_train_poly)
    mse_train = mean_squared_error(y_train, y_pred_train)
    mse_degrees_train.append(mse_train)
    
plt.plot(degrees, mse_degrees_train, 'b', label='Training set')
plt.xlabel("Degree of polynomial")
plt.ylabel("MSE")
plt.legend()
plt.grid()
plt.show()

best_degree = 1 + mse_degrees_train.index(min(mse_degrees_train))
print(f'The best performance is {min(mse_degrees_train)} with degree {best_degree}.')
best_degrees.append(best_degree)
```

### (2)

```python
x_train_smaller, x_validation, y_train_smaller, y_validation = train_test_split(x_train, y_train, test_size=0.5, random_state=42)

mse_degrees_validation = []
degrees = [i for i in range(1, 6)]

for d in degrees:
    poly = preprocessing.PolynomialFeatures(degree=d)
    poly.fit(x_train_smaller)
    x_train_smaller_poly = poly.transform(x_train_smaller)
    x_validation_poly = poly.transform(x_validation)
    
    model = LinearRegression()
    model.fit(x_train_smaller_poly, y_train_smaller)
    
    
    y_pred = model.predict(x_validation_poly)
    mse_validation = mean_squared_error(y_validation, y_pred)
    mse_degrees_validation.append(mse_validation)

plt.plot(degrees, mse_degrees_validation, 'r', label='Validation set')
plt.xlabel("Degree of polynomial")
plt.ylabel("MSE")
plt.legend()
plt.grid()
plt.yscale("log")
plt.show()

best_degree = 1 + mse_degrees_validation.index(min(mse_degrees_validation))
print(f'The best performance is {min(mse_degrees_validation)} with degree {best_degree}.')
best_degrees.append(best_degree)
```

The approaches yield consistent results regarding the optimal degree, which is 1.
