# Task 5: Support Vector Regression


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
np.random.seed(42)
x = np.sort(5 * np.random.rand(200, 1), axis=0)
y = np.cos(x).ravel() + 0.2 * np.random.normal(size=200)

epsilons = [0, 0.1, 0.5, 1.0, 3.0]
colors = ['g', 'b', 'r', 'orange', 'purple']

plt.figure(figsize=(10, 8))
plt.scatter(x, y, color='k', label='data', s=10)

for epsilon, color in zip(epsilons, colors):
    
    model = SVR(kernel='rbf', epsilon=epsilon)
    y_pred = model.fit(x, y).predict(x)
    
    plt.plot(x, y_pred, label=f'SVR (ε={epsilon})', color=color)
    #plt.plot(x, y_pred - epsilon, color=color, linestyle='dashed')
    #plt.plot(x, y_pred + epsilon, color=color, linestyle='dashed')
    

plt.title('Support Vector Regression with RBF Kernel')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x, y, color='k', label='data', s=10)

for epsilon, color in zip(epsilons, colors):
    
    model = SVR(kernel='rbf', epsilon=epsilon)
    y_pred = model.fit(x, y).predict(x)
    
    plt.plot(x, y_pred, label=f'SVR (ε={epsilon})', color=color)
    plt.plot(x, y_pred - epsilon, color=color, linestyle='dashed')
    plt.plot(x, y_pred + epsilon, color=color, linestyle='dashed')
    

plt.title('Support Vector Regression with RBF Kernel')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
