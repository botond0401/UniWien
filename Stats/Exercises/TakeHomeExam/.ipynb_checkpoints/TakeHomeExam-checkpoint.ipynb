### b)
Consider the US crime data set but use only the variables “PctKids2Par”, “NumStreet” and “racePctWhite” as well as an intercept column. Implement an appropriate MCMC algorithm that allows you to sample from the d = 4 dimensional posterior density.

```python
# importing the dataset and the column names
df_columns = pd.read_csv('columns.csv', header=None)
columns = df_columns[0].to_list()

df_communities = pd.read_csv('communities.csv', header=None)
df_communities = df_communities.iloc[:, 5:]
df_communities.columns = columns
df_communities = df_communities.reset_index(drop=True)
```

```python
# building our target and explanatory variables
target_column = columns[-1]
# these 3 columns do not have invalid entries, such as missing values or a "?" character
explanatory_columns = ['PctKids2Par', 'NumStreet', 'racePctWhite']
df_y = df_communities[target_column].astype(float)
df_x = df_communities[explanatory_columns].astype(float)
# adding an intercept as required
df_x['intercept'] = 1
x = np.array(df_x).astype(float)
y = np.array(df_y).astype(float)
```

```python
def f(β: np.ndarray, λ = 1.) -> float:
    """
    The posterior density function.
    
    Parameters:
    - β (np.ndarray): The values for the parameter β
    - λ (np.array): Tuning parameter

    Returns:
    - float: A float between 0 and 1 representing the probability of β being the true parameter
    """
    # the Laplace prior
    p = np.prod(np.exp(- λ * np.abs(β)))
    # the likelihood part
    p *= np.prod(np.exp(- 1 / 2 * (y - x @ β) ** 2))
    return p

def q(β: np.ndarray, sigma: float = .1) -> np.ndarray:
    """
    A (naive) Gaussian proposal density funuction.
    
    Parameters:
    - β (np.ndarray): The values for the parameter β
    - sigma (float): The standard deviation of the Gaussian distribution for the new values

    Returns:
    - np.ndarray: New values for the parameter β
    """
    d = len(β)
    return β + sigma * np.random.randn(d)
```

```python
def create_MC(
    q: Callable[[np.ndarray], np.ndarray],
    λ: float = 1.,
    initial_beta: np.ndarray = np.array([0, 0, 0, 0]),
    t: int = 100
) -> np.ndarray:
    """
    Generate a Markov chain using the Metropolis-Hastings algorithm.

    Parameters:
    - q (callable): A proposal function
    - initial_beta (np.ndarray): Initial value for the β variable
    - t (int): Number of iterations

    Returns:
    - np.ndarray: An array representing the generated Markov chain for β
    """
    β = np.zeros((t, initial_beta.shape[0]))
    β[0] = initial_beta
    for i in range(1, t):
        z = q(β[i-1])
        # using only symmetric proposition distrbutions, so q can be omitted
        α = min(1, f(z, λ) / f(β[i-1], λ))
        rand_num = random.random()
        β[i] = z if rand_num < α else β[i-1]
    return β
```

```python
# visualizing a generated Markov chain with length 100
MC = create_MC(q)
plt.figure(figsize=(10, 6))

# Plot each row as a separate line
for i in range(MC.shape[1]):
    plt.plot(MC[:, i], label=f'Parameter {i + 1}')

# Add labels and legend
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.title('The generated Markov chains')
plt.legend(loc='best')
```

### c)
For each parameter βj , j = 1, . . . , 4, (numerically) compute a (one-dimensional) 95% Bayesian credible interval. How do these intervals change as λ changes? Provide some visualization.

```python
# we have 4 parameters
len_beta = 4
# values for lamdba
lambda_values = [0, 1, 10, 50, 100]
sample_size = 10000
samples = {}
# creating a dictionary with lamdbas as keys and the corresponding Markov chains as values
for lambda_value in lambda_values:
    samples[lambda_value] = create_MC(q, λ=lambda_value, t=sample_size)
    
# Function to compute credible interval
def compute_credible_interval(samples: np.ndarray, α: float = 0.05) -> Tuple[float, float]:
    """
    Calculate the credible interval of a dataset for a given α.

    Parameters:
    - samples (np.ndarray): A dataset
    - α (float): The significance level

    Returns:
    - float, float: The lower and upper credibel limits
    """
    lower_bound = np.percentile(samples, α / 2 * 100)
    upper_bound = np.percentile(samples, (1 - α / 2) * 100)
    return lower_bound, upper_bound


fig, axes = plt.subplots(nrows=4, ncols=1, figsize=(8, 12))
# plotting how the credible intervals change with respect to λ for each parameter
for i in range(len_beta):
    x_coordinate = 1
    for lambda_value, sample in samples.items():
        sample = sample[:, i]
        lower, upper = compute_credible_interval(sample)
        axes[i].plot([x_coordinate, x_coordinate], [lower, upper],
                     marker='o')
        x_coordinate += 1
    axes[i].set_title(f'β{i+1} Credible Intervals')
    axes[i].set_xticks(range(1, len(lambda_values) + 1))
    axes[i].set_xticklabels(lambda_values)

    axes[i].set_xlabel('λ')
plt.tight_layout()
plt.show()
```

As expected one can observe that the parameters approach 0 as λ increases.

## Exercise 2
Design a pivotal bootstrap type resampling scheme for constructing confidence intervals for population graph totals in statistical network analysis.
### a)
Implement it on the Facebook data set from the lecture to obtain a bootstrap confidence interval for the population proportion of TV-politics edges and plot the bootstrap distribution of your estimator for just a handful of different random samples. Pick any graph sampling design you consider reasonable. Proceed as in the lecture, assuming that you don’t have access to neither E nor N, but that you can only retrieve edge information
once you have sampled one of its incident vertices.


```python
# importing our data
edges = pd.read_csv('musae_facebook_edges.csv')
nodes = pd.read_csv('musae_facebook_target.csv')

# create a graph using the networkx package
G = nx.from_pandas_edgelist(edges, 'id_1', 'id_2')

# for each node adding attribute 'page_type' from the data
for _, row in nodes.iterrows():
    node = row['id']
    G.nodes[node]['page_type'] = row['page_type']
    
# using the node attributes adding attribute 'is_relevant' to the edges
# this attribute is 1 if its nodes are 'politician' and 'tvshow', otherwise 0
for edge in G.edges:
    val1 = G.nodes[edge[0]]['page_type']
    val2 = G.nodes[edge[1]]['page_type']
    if ((val1 == 'tvshow') & (val2 == 'politician')) | ((val1 == 'politician') & (val2 == 'tvshow')):
        G.edges[edge]['is_relevant'] = 1
    else:
        G.edges[edge]['is_relevant'] = 0
        
```

```python
def random_sample(G, n: int = 1000):
    """
    Perform induced subgraph sampling.

    Parameters:
    - G (NetworkX graph): The input graph
    - n (int): Number of nodes to sample, default is 1000

    Returns:
    - NetworkX graph: A subgraph of the input graph
    """
    V = random.sample(G.nodes, k=n)
    G_sample = G.subgraph(V).copy()
    for edge in G_sample.edges:
        G_sample.edges[edge]['is_relevant'] = G.edges[edge]['is_relevant']

    return G_sample


def TVP_proportion(G) -> float:
    """
    Calculate the proportion of 'is_relevant' edges in a graph.

    Parameters:
    - G (NetworkX graph): The input graph

    Returns:
    - float: The proportion of edges labeled as 'is_relevant' in the graph
    """
    num_edges_all = G.number_of_edges()
    num_edges_TVP = sum(1 for edge in G.edges() if G.edges[edge]['is_relevant'] == 1)
    return num_edges_TVP / num_edges_all


def pivotal_bootstrap(G, B: int = 100, α: float = 0.05) -> Tuple[float, float, List[float]]:
    """
    Perform pivotal bootstrap to construct a credible interval for the proportion
    of 'is_relevant' edges in a graph

    Parameters:
    - G (NetworkX graph): The input graph
    - B (int): Number of bootstrap samples
    - α (float): Significance level for confidence interval, default is 0.05

    Returns:
    - float, float, List[float]: The lower and upper bounds of the confcredibleidence interval,
      and the list of proportion values from the bootstrap samples
    """
    e = G.number_of_edges()
    mean = TVP_proportion(G)
    sample_values = []
    proportion_values = []
    for _ in range(B):
        # we sample with replacement from the edges
        edges = random.choices(list(G.edges), k=e)
        # if an edge is chosen more than once, it is taken into account more than once
        is_relevant = [G.edges[edge]['is_relevant'] for edge in edges]
        proportion = sum(is_relevant) / e
        sample_values.append(proportion - mean)
        proportion_values.append(proportion)
    lower_quantile = np.percentile(sample_values, 100 * (1 - α / 2))
    upper_quantile = np.percentile(sample_values, 100 * (α / 2))
    return mean - lower_quantile, mean - upper_quantile, proportion_values
```

```python
# running our bootstrap function for some saomple graphs
prop_true = TVP_proportion(G)
num_samples = 4
sample_size = 1000

# Create a grid of subplots
num_columns = 2
num_rows = (num_samples + 1) // 2 
fig, axs = plt.subplots(num_rows, num_columns, figsize=(16, 8))

# Flatten the axs array for easier indexing in the loop
axs = axs.flatten()

# creating a histogram for each sample
for i in range(num_samples):

    G_sample = random_sample(G=G, n=sample_size)
    prop_sample = TVP_proportion(G_sample)

    l, u, proportion_values = pivotal_bootstrap(G=G_sample, B=1000)
    axs[i].hist(proportion_values, bins=15, color='blue', edgecolor='black')
    axs[i].set_title(f'{i+1}. subgraph')
    axs[i].set_xlabel('Estimator')
    axs[i].set_ylabel('Frequency')
    axs[i].axvline(prop_true, color='green', linestyle='-', linewidth=2, label='True proportion')
    axs[i].axvline(prop_sample, color='orange', linestyle='-', linewidth=2, label="Subgraph's proportion")
    axs[i].axvline(l, color='red', linestyle='--', linewidth=2, label='Credible limits')
    axs[i].axvline(u, color='red', linestyle='--', linewidth=2)
    axs[i].legend(loc='best')
    
plt.tight_layout()
```

### b)
Verify the validity of your method by running a Montecarlo simulation, recalling that
you know the true population proportion since the population graph is given.

```python
def monte_carlo(
    G,
    α: float = 0.05,
    n: int = 1000,
    B: int = 100,
    iters: int = 1000
) -> float:
    """
    Perform Monte Carlo simulation to estimate the coverage probability of a pivotal
    bootstrap confidence interval for the proportion of 'is_relevant' edges in a graph.

    Parameters:
    - G (NetworkX graph): The input graph
    - α (float): Significance level for pivotal bootstrap confidence interval, default is 0.05
    - n (int): Number of nodes to sample in each Monte Carlo iteration, default is 1000
    - B (int): Number of bootstrap samples in each pivotal bootstrap iteration, default is 100
    - iters (int): Number of Monte Carlo iterations, default is 1000

    Returns:
    - float: The estimated coverage probability of the pivotal bootstrap confidence interval
    """
    prop_true = TVP_proportion(G=G)
    valid = 0
    for _ in range(iters):
        G_sample = random_sample(G=G, n=n)

        l, u, _ = pivotal_bootstrap(G=G_sample, B=B)
        if (l <= prop_true) & (u >= prop_true):
            valid += 1
    return valid / iters
```

```python
# running the MC simulation for different sample sizes and plotting the results
ratios = []
# number of initial nodes to choose for the subgraph
sample_sizes = [500, 1000, 2000, 4000]
for n in sample_sizes:
    ratios.append(monte_carlo(G, n=n))

# Create a line plot
plt.plot(ratios, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.xlabel('Sample size')
plt.ylabel('Coverage probability')
plt.title('Coverage probability with respect to the sample size')

# Show the plot
plt.show()
```

Unfortunately the results do not reach 95% as expected, but this can be imporved by further increasing the sample size or using a more advanced graph sampling design.
