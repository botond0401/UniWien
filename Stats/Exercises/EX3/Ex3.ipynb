```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
def plot_dist(x, theta):
    n = len(x)
    x_sum = np.sum(x)
    
    a = x_sum + 1
    b = n - x_sum + 1

    fig, axs = plt.subplots(1, 2, figsize=(10, 4))

    axs[0].hist(x, bins=[-0.5, 0.5, 1.5], edgecolor='black', align='mid', rwidth=0.8)
    axs[0].set_xlabel('Values')
    axs[0].set_ylabel('Frequency')
    axs[0].set_xticks([0, 1])
    axs[0].set_title(f'Histogram of values sampled with theta: {theta}')

    values = np.linspace(0, 1, 1000)
    pdf_values = beta.pdf(values, a, b)
    axs[1].plot(values, pdf_values, label=f'Beta({a}, {b})')
    axs[1].set_xlabel('Probability of Success (p)')
    axs[1].set_ylabel('Probability Density')
    axs[1].legend()
    axs[1].set_title(f'Posterior Distribution with Sample Size {n}')

    plt.tight_layout()
    
```

```python
for _ in range(4):
    theta = np.random.rand()
    for n in [100, 500, 1000, 5000]:
        x = np.random.choice([0, 1], size=n, p=[1-theta, theta])
        plot_dist(x, theta)
```

# Exercise 3.2

```python
def h(x, y):
    return np.sin(x * y)**2 * np.cos(-1 * x * y)**2 * np.exp(-8 * x**2 - 2 * np.abs(y)**3)
```

```python
def generate_points(T, Y):
    n = len(Y)
    U = np.random.rand(n)
    X = []
    for i in range(n):
        if U[i] <= T(Y[i][0], Y[i][1]):
            X.append(Y[i])
    return X
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def plot_points(points):
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)

    X, Y = np.meshgrid(x, y)
    Z = h(X, Y)

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    ax.plot_surface(X, Y, Z, cmap='viridis')

    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('value of h(x, y)')
    ax.set_title('Plot of h(x, y) with generated points')
    
    x_values, y_values = zip(*points)
    z_values = [h(x, y) for x, y in points]
    scatter = ax.scatter(x_values, y_values, z_values, c='red', cmap='viridis', marker='o', label='Generated points')
    ax.legend()
    plt.show()
```

## a)

```python
def g_a(x, y):
    return np.exp((-1 / 2) * (x**2 + y**2)) / (2*np.pi)
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = h(X, Y)/g_a(X, Y)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set the initial view angles
ax.view_init(elev=20, azim=30)  # Adjust these angles as needed

# Show the plot
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot()
ax.plot(X, Z)
ax.set_ylim(bottom=0, top=0.1)

plt.show()
```

```python
M_a = 0.1
```

```python
def T_a(x, y):
    if g_a(x, y) > 0:
        return h(x,y) / (M_a * g_a(x, y))
    else:
        return 0.0
    
```

```python
n = 1000
mean = [0, 0]
covariance_matrix = [[1, 0], [0, 1]] 
Y_a = np.random.multivariate_normal(mean, covariance_matrix, n)
```

```python
X_a = generate_points(T_a, Y_a)
len(X_a)
```

```python
plot_points(X_a)
```

## c)

```python
a_ratios = []
b_ratios = []
for e in range(3, 7):
    
    n = 10**e
    
    mean = [0, 0]
    covariance_matrix = [[1, 0], [0, 1]] 
    Y_a = np.random.multivariate_normal(mean, covariance_matrix, n)
    X_a = generate_points(T_a, Y_a)
    a_ratios.append(int(n / len(X_a)))
    
    y_b_0 = np.random.uniform(-1, 1, n)
    y_b_1 = np.random.uniform(-1, 1, n)
    Y_b = list(zip(y_b_0, y_b_1))
    X_b = generate_points(T_b, Y_b)
    b_ratios.append(n / len(X_b))
```

```python
a_avg = sum(a_ratios) / len(a_ratios)
b_avg = sum(b_ratios) / len(b_ratios)
print(f'In the first case for sample size N on average we need {a_avg} * N iterations.')
print(f'In the second case for sample size N on average we need {b_avg} * N iterations.')
```
