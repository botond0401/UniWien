**Goal:**
$$\min_x f(x) := \frac{1}{m}\log\lrb{\sum_{i=1}^m e^{a_i^\top x}} + \frac{\mu}{2}\n{x}^2.$$

**Algorithm:** $$x_{k+1} = x_k - [\nabla^2 f(x_k)]^{-1} \nabla f(x_k)$$

## Ingredients for Netwon method

For Newton method, we need to compute its gradient and Hessian.

$$f(x) = \frac 1 m\ell(Ax) + \frac{\mu}{2}\n{x}^2, \qquad \ell(z) = \mathrm{logsumexp}(z)$$

Gradient of $\ell$ is easy:
$$\nabla \ell(z) = \mathrm{softmax}(z) = \lrb{\frac{e^{z_i}}{\sum e^{z_j}}}_{i=1}^m =: \sigma(x)$$

Hence,
$$\nabla f(x) = \frac 1m A^\top \nabla \ell(Ax) +\mu x =\frac 1 m A^\top
\sigma(Ax) + \mu x$$

For the Hessian, we also apply the chain rule:
$$\nabla^2(\phi(Ax)) =A^\top \nabla^2\phi(Ax) A$$

Hessian of $\sigma$ is also easy (homework):
$$\nabla^2 \sigma(x) = \mathop{\mathrm{diag}}(\sigma(x)) - \sigma(x)\sigma(x)^\top.$$

Altogether,
\begin{align}
\nabla f(x) &= \frac 1m A^\top \s(Ax) +  \mu x\\
\nabla^2 f(x) &= \frac 1 m A^\top [\diag(\s(Ax)) - \s(Ax)\s(Ax)^\top] A + \mu I
\end{align}

The function $f$ is strongly convex, thus its Hessian is positive
definite (hence invertible).

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as LA
import scipy.optimize as opt
from scipy.special import logsumexp, softmax
```

We define data and all the ingredients for the algorithm

```python
m = 200
n = 100
mu = 1.0e-1
A = np.random.randn(m, n)

def f(x):
    return 1. / m * logsumexp(A @ x) + mu/2 * LA.norm(x)**2

def grad_f(x):
    return 1. / m *  A.T @ softmax(A @ x) + mu * x

def hess_f(x):
    s = softmax(A @ x)
    return 1. / m * A.T @ (np.diag(s) - np.outer(s, s)) @ A + mu * np.eye(n)
```

## Checking with automatic differentiation

```python
import jax.numpy as jnp
import jax.numpy.linalg as LA_jax
from jax import grad, hessian
from jax.scipy.special import logsumexp as logsumexp_jax
from jax.config import config; config.update("jax_enable_x64", True)


def f_jax(x):
    return 1. / m * logsumexp_jax(A @ x) + mu/2 * LA_jax.norm(x)**2


u = np.random.randn(n)
grad_jax = grad(f_jax)
hess_jax = hessian(f_jax)
print("Gradients: ", LA.norm(grad_jax(u) - grad_f(u)))
print("Hessians: ", LA.norm(hess_jax(u) - hess_f(u)))
```

## Better code

How to make the above code more suitable for Newton’s method?

```python
def oracle_f(x):
    Ax = A @ x
    fx = 1. / m * logsumexp(Ax) + mu/2 * LA.norm(x)**2
    s = softmax(Ax)
    g = 1. / m *  A.T @ s + mu * x
    H = 1. / m * A.T @ (np.diag(s) - np.outer(s, s)) @ A + mu * np.eye(n)
    return fx, g, H
```

Now, let’s implement the algorithm

```python
def newton_method(oracle_f, x0, max_iter=100, tol=1e-6):
    x = x0.copy()
    residuals = []
    energy = []
    for i in range(max_iter):
        fx, g, H = oracle_f(x)
        res = LA.norm(g)
        residuals.append(res)
        energy.append(fx)
        if res < tol:
            print(f"Newton method converged in {i+1} iterations")
            break
        else:
            d = LA.solve(H, g, assume_a='pos')
            x -= d
            
    return x, residuals, energy
```

## Solving and plotting

Finally, let’s solve our problem

```python
x0 = np.ones(n)
MAX = 20
%time ans1 = newton_method(oracle_f, x0, max_iter=MAX, tol=1.0e-15)

plt.plot(ans1[1], label="Newton")
plt.legend()
plt.xlabel("iterations")
plt.ylabel("residual")
plt.yscale("log")
```

## Approximating $\sqrt{a}$

Write a function that approximately computes $\sqrt{a}$. In other words,
you want to solve equation $x^2 = a$.

$$x_{k+1} = x_k - \frac{x_k^2 - a}{2x_k} = \frac{1}{2} \left(x_k + \frac{a}{x_k}\right)$$

## Newton method and fractals (Julia set)

1.  We are given a polynomial $p(z)\in \mathbb{C}[z]$ (with complex
    coeficients) of the degree $m$.

2.  This polynomial has $m$ roots $\xi_1,\dots, \xi_m$, that is
    $p(\xi_i) = 0$. We know that one can apply the Newton method
    $$z_{k+1}=z_k
    -\frac{p(z_k)}{p'(z_k)}$$ to approximate $\xi_i$.

    In other words, although we work now in the complex plain, the
    scheme looks exactly the same.

3.  We know that convergence of Newton’s method depends on the initial
    point $z_0$: sometimes it will converge, sometimes not. This is what
    we want to study.

4.  In a certain region, say $[-1,1]\times [-1,1]$, we want to examine
    as many points as we can to see whether the Newton method diverges
    or not and if not, to which root it converges. Thus, we can put all
    our $z_0$ into $m+1$ classes: one for each root and one for
    divergence. We can color all these classes in different colors.

5.  This is how we get the Newton fractal. More details on
    [3blue1brown](https://www.youtube.com/watch?v=-RdOwhmqP5s)



## Gauss-Newton method

Suppose we want to solve a nonlinear least squares problem
$$\min F(x):= \frac 1 2\n{f(x)}^2 = \frac 12 \sum_{i=1}^m f_i(x)^2.$$

We want to use Newton’s method for this problem. What is the Hessian of
$F$? It is given by
$$\nabla^2 F(x) = \sum_i \nabla f_i(x) \nabla f_i(x)^\top + \sum_{i} f_i(x)
\nabla^2 f_i(x).$$

If we suppose that the Newton method converges, then each $f_i$ becomes
smaller and smaller, hence it makes sense to just ignore it and instead
of the standard Newton direction
$$\text{Newton:}\qquad d_k = [\nabla^2 F(x)]^{-1} \nabla F(x_k)$$
consider its approximated (and cheaper! version
$$\text{Gauss-Newton:}\qquad d_k = \left[\sum_i \nabla f_i(x) \nabla f_i(x)^\top \right]^{-1} \nabla
F(x_k),$$

This update is cheaper because it doesn’t require us to compute Hessians
of $f_i$; we only use their gradients.
