```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    n = len(x)
    return 100 * np.sum((x[1:] - x[:-1]**2)**2) + (1 - x[0])**2

def gradient(x):
    n = len(x)
    grad = np.zeros_like(x)
    grad[0] = 2 * x[0] - 2 - 400 * x[0] * x[1] + 400 * x[0]**3
    for i in range(1, n-1):
        grad[i] = 200 * x[i] - x[i-1]**2 - 400 * x[i] * x[i+1] + 400 * x[i]**2
    grad[n-1] = 200 * (x[n-1] - x[n-2]**2)
    return grad

def gradient_descent(x0, alpha, num_iterations):
    x = x0.copy()
    trajectory = [x.copy()]

    for _ in range(num_iterations):
        x = x - alpha * gradient(x)
        trajectory.append(x.copy())

    return np.array(trajectory)

# Set parameters
n = 5  # You can choose any value for n
learning_rate = 0.0001
iterations = 1000


def plot_gd(initial_point):
    # Pick and print starting values
    print(f"X0 is {initial_point}")

    # Run gradient descent
    traj = gradient_descent(initial_point, learning_rate, iterations)

    final_x = traj[-1]
    # Plot convergence trajectory
    plt.figure(figsize=(10, 6))
    plt.plot(traj[:, 0], label='x[0]')
    plt.plot(traj[:, 1], label='x[1]')
    plt.plot(traj[:, 2], label='x[2]')
    plt.plot(traj[:, 3], label='x[3]')
    plt.plot(traj[:, 4], label='x[4]')
    plt.title('The value of the variables during GD')
    plt.xlabel('Iterations')
    plt.ylabel('Variable Value')
    plt.legend()
    plt.show()

    # Print final results
    print("Final value of f(x): ", func(final_x))
```

```python
initial_points = [1.1, 1.5, 2, 3]
for initial_point in initial_points:
    initial_point = initial_point*np.ones(n)
    plot_gd(initial_point)
```

```python
initial_point = np.array([0, 1, 1, 1, 0.5])
plot_gd(initial_point)
```

```python
# Define the function
def f(x1, x2):
    return np.sqrt(x1**2 + x2**2) + np.sin(4*np.arctan(x2/x1))

# Generate points for the plotting
x1 = np.linspace(-3, 3, 100)
x2 = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Plot level sets
plt.figure(figsize=(15, 8))
plt.contour(X1, X2, Z, levels=20)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Level Sets of f(x1, x2)')
plt.grid(True)
plt.show()
```

The function is neither differentiable, nor convex. It is not convex, because we can observe that there sections connecting two points on the same plot level, which have parts on larger plot levels, and this cannot happen with a convex function. It appears to be not differentiable, because the level sets close to (0, 0) have sharp corners.
