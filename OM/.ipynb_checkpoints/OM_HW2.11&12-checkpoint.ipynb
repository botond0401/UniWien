```python
import numpy as np
import matplotlib.pyplot as plt
import autograd.numpy as np
from autograd import grad
```

## **Problem 11.**
Implementing a function, which solves our problem using gradient descent for this given function, and creates a required plots with respect to the given number of points(N).

```python
def solve_and_plot(N, learning_rate=0.001, epochs=1000, seed=42):
    M = 2*N+1

    # Our function to be approximated. We pretend that we don't know f
    def f(x):
        return np.sin(x * np.sqrt(2)) + np.sin(x**2)

    s = np.arange(M) * 2 * np.pi / M # sample points
    y = f(s) # function values at sample points

    # Constructing A
    A = np.zeros((M, 2*N + 1))  
    A[:, 0] = 1 # Initialize the matrix with ones in the first column

    for i in range(1, N + 1):
        A[:, i] = np.cos(i * s)
        A[:, N + i] = np.sin(i * s)

    # Solve \min_x |Ax-y|^2 - our objective function
    def F(x):
        return np.linalg.norm(A @ x - y)**2

    # our objective function's gradient
    grad_F = grad(F)

    # implementing gradient descent
    def gradient_descent(learning_rate=learning_rate, epochs=epochs, seed=seed):
        np.random.seed(seed)

        # initial guess for the center (x, y) and radius r
        x = np.random.rand(M)

        for epoch in range(epochs):
            grad = grad_F(x)
            x -= learning_rate * grad
            # print the objective function value every 100 epochs
            if epoch % 100 == 0:
                cost = F(x)
                print(f"Epoch {epoch}, Cost: {cost}")

        return x

    # getting our x
    x = gradient_descent(learning_rate, epochs, seed)

    # Given x, write a function that computes a trigonometric polynomial p
    def p(t):
        P = x[0]
        for i in range(1, N + 1):
            P += x[i] * np.cos(i * t) + x[N + i] * np.sin(i * t)
        return P

    # Make all plots
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

    axs[0].scatter(s, np.abs(f(s) - p(s)))
    axs[0].set_title('Plotting the residuals')
    axs[0].set_xlabel('x')
    axs[0].set_ylabel('|f(x) - p(x)|')
    
    # For plotting f and p use a dense grid, like 
    s_dense = np.linspace(0,2*np.pi,1000)
    axs[1].plot(s_dense, f(s_dense), label="original function")
    axs[1].plot(s_dense, p(s_dense), label="reconstructed function")
    axs[1].plot(s, f(s), ".r", label="sample points")
    axs[1].set_title('Plotting the function')
    axs[1].legend(loc='lower left')
    axs[1].set_xlabel('x')
    axs[1].set_ylabel('function value at x')
```

### N = 10

```python
solve_and_plot(10)
```

### N = 100

```python
x = solve_and_plot(100)
```

The method already approximates really well our function with N being 10, but after setting N to 100, it is basically perfectly fitted and the maximal absolute residual also increased from above 3 to 2.

## **Problem 12.**
Implementing Kaczmarz's method.

```python
def kaczmarz(A, b, x0, iterations=10000):
    x = x0.copy()
    m, _ = A.shape

    objective_values = []
    for i in range(iterations):
        i %=m
        ai = A[i, :]
        alpha = np.dot(ai, x) - b[i]
        denominator = np.dot(ai, ai)
        x -= (alpha / denominator) * ai

        objective_value = np.linalg.norm(A @ x - b)**2
        objective_values.append(objective_value)

    return x, objective_values
```

Implementing randomized Kaczmarz's method.

```python
def randomized_kaczmarz(A, b, x0, iterations=10000):
    x = x0.copy()
    m, _ = A.shape

    objective_values = []
    for _ in range(iterations):
        i = np.random.choice(m, p=np.linalg.norm(A, axis=1)**2 / np.linalg.norm(A)**2)
        ai = A[i, :]
        alpha = np.dot(ai, x) - b[i]
        denominator = np.dot(ai, ai)
        x -= (alpha / denominator) * ai

        objective_value = np.linalg.norm(A @ x - b)**2
        objective_values.append(objective_value)

    return x, objective_values
```

Implementing a function, which runs both methods for a given x0 and compares them with plots of the costs along the iterations.

```python
def compare_kaczmarz(A, b, x0, iterations=10000):
    
    _, objective_values_kaczmarz = kaczmarz(A, b, x0, iterations)
    _, objective_values_randomized = randomized_kaczmarz(A, b, x0, iterations)   
    
    print(f'The final value of the objective function with Kaczmarz\'s method: {objective_values_kaczmarz[-1]}')
    print(f'The final value of the objective function with randomized Kaczmarz\'s method: {objective_values_randomized[-1]}')
    
    # plot the objective values in each iteration
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
    
    x_values = list(range(1, iterations+1))
    
    axs[0].plot(x_values, objective_values_kaczmarz)
    axs[0].set_title('Performance with Kaczmarz\'s method')
    axs[0].set_xlabel('iteration')
    axs[0].set_ylabel('log of objective value')
    axs[0].set_yscale('log')  # Set y-axis to logarithmic scale
    
    axs[1].plot(x_values, objective_values_randomized)
    axs[1].set_title('Performance with randomized Kaczmarz\'s method')
    axs[1].set_xlabel('iteration')
    axs[1].set_ylabel('log of objective value')
    axs[1].set_yscale('log')  # Set y-axis to logarithmic scale
```

The randomized algorithm works better in most cases:

```python
np.random.seed(42)
m, n = 100, 100
A = np.random.rand(m, n)
x_true = np.random.rand(n)
b = A @ x_true + 0.1 * np.random.rand(m)

# Initial guess for x
x0 = np.zeros(n)

# Solve using Kaczmarz method
compare_kaczmarz(A, b, x0, 10000)
```

But if we have a well-conditioned A matrix and create b after creating x, the basic Kaczmarz's method works better:

```python
# Create a well-conditioned matrix A
A = np.random.rand(100, 100)

# Ensure that A is well-conditioned
A = A @ A.T

# Create a corresponding b and x_true
x_true = np.random.rand(100)
b = A @ x_true

# Initial guess for x
x0 = np.zeros(100)

# Solve using Kaczmarz method
compare_kaczmarz(A, b, x0, 100000)
```

Implementing gradient descent for this type of functions.

```python
def gradient_descent(A, b, x0, learning_rate=0.01, iterations=1000):
    x = x0.copy()

    objective_values = []
    for _ in range(iterations):
        gradient = 2 * A.T @ (A @ x - b)
        x -= learning_rate * gradient

        objective_value = np.linalg.norm(A @ x - b)**2
        objective_values.append(objective_value)
            
    return x, objective_values
```

Implementing a function, which runs GD and randomized Kacmarz's method for a given x0, and compares their performance with plots of the costs with respect to the number of iterations.

```python
def compare_with_GD(A, b, x0, learning_rate=0.01, iterations=100):
    
    #  one iteration of Kaczmarz method is roughly m times cheaper than the one of GD.
    m, _ = A.shape
    iterations_kaczmarz = m * iterations
    _, objective_values_GD = gradient_descent(A, b, x0, learning_rate, iterations)
    _, objective_values_randomized = randomized_kaczmarz(A, b, x0, iterations_kaczmarz)   
    
    print(f'The final value of the objective function with gradient descent: {objective_values_GD[-1]}')
    print(f'The final value of the objective function with randomized Kaczmarz\'s method: {objective_values_randomized[-1]}')
    
    # plot the objective values in each iteration
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
    
    x_values_GD = list(range(1, iterations+1))
    x_values_kaczmarz = list(range(1, iterations_kaczmarz+1))
    
    axs[0].plot(x_values_GD, objective_values_GD)
    axs[0].set_title('Performance with gradient descent')
    axs[0].set_xlabel('iteration')
    axs[0].set_ylabel('log of objective value')
    axs[0].set_yscale('log')  # Set y-axis to logarithmic scale
    
    axs[1].plot(x_values_kaczmarz, objective_values_randomized)
    axs[1].set_title('Performance with randomized Kaczmarz\'s method')
    axs[1].set_xlabel('iteration')
    axs[1].set_ylabel('log of objective value')
    axs[1].set_yscale('log')  # Set y-axis to logarithmic scale
```

When comparing randomized Kaczmarz's method with gradient descent we allow Kaczmarz's method to have m times more iterations as one of its steps is roughly m times cheaper than the one of GD. 

```python
np.random.seed(42)
m, n = 100, 100
A = np.random.rand(m, n)
x_true = np.random.rand(n)
b = A @ x_true + 0.1 * np.random.rand(m)

# Initial guess for x
x0 = np.zeros(n)

# Solve using Kaczmarz method
compare_with_GD(A, b, x0, 0.0001, 1000)
```

Meaning the Kaczmarz's method works much better if they run for the same amount of time.
